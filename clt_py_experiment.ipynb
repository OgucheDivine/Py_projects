##### Algorithm
based on the video the numbers all started from the centre (0),and through several layers of forward, backward and same movement (-1 ,0,1) which are random,they will be a central tendency.<br>
so
1. I will create the baseline with np.zeros
2. Generate the diff layers with np.random.randint
3. combine 
4. plot graph
5. put all in a func

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#step 1
baseline=np.zeros(1000)
baseline
```

```python
#step2
#layer1
layer_1=np.random.randint(-1,2,size=baseline.shape)
#layer2
layer_2=np.random.randint(-1,2,size=baseline.shape)
#layer3
layer_3=np.random.randint(-1,2,size=baseline.shape)
#layer4
layer_4=np.random.randint(-1,2,size=baseline.shape)
#layer5
layer_5=np.random.randint(-1,2,size=baseline.shape)

end_result=baseline+layer_1+layer_2+layer_3+layer_4+layer_5
values,count=np.unique(end_result,return_counts=True)
result=pd.Series(count,index=values)
result.sort_values(ascending=False)
#actually it worksðŸ˜­ðŸ˜­
```

```python
#step 3
fig,axs=plt.subplots(1,2,figsize=(19,9))

result.plot(label='Frequency',ylabel='Frequency',xlabel='Values',title='Frequency of Values after 5 layers',ax=axs[0])
result.plot(kind='bar',ylabel='Frequency',xlabel='Values',title='Frequency of Values after 5 layers',ax=axs[1])
#I thought it wasn't trueðŸ¤¦â€â™‚ï¸
#so i want to  push all this into a func and add a couple of pseudo random gen for reproducibilty
```

```python
#step 5
def pass_zeros_arr_into_levels(levels,size=1000):
    #creating base arr
    base=np.zeros(size)
    #creating and adding levels
    for i in range(levels):
        rng=np.random.default_rng(i)
        level=rng.integers(-1,2,size)
        base+=level
    #creating pd series object which hold the values and counts
    values,count=np.unique(base,return_counts=True)
    result=pd.Series(count,index=values)
    return result
pass_zeros_arr_into_levels(5)
```

```python
#Now let's plot a couple of graphs to show relationship

#Value count at level 10,20,30 and 40

fig,axs=plt.subplots(1,2,figsize=(19,9))
lvl_10_count=pass_zeros_arr_into_levels(10)
lvl_10_count.plot(ylabel='Frequency',xlabel='Values',title='Frequency of Values after 10 layers',ax=axs[0])
lvl_10_count.plot(kind='bar',ylabel='Frequency',xlabel='Values',title='Frequency of Values after 10 layers',ax=axs[1])
```
